```python
import pandas as pd
import numpy as np
```

```python
# Student data filepath
student_data_path = "data/StudentPerformanceFactors.csv"

# Student data dataframe created by reading the CSV with pandas
student_df = pd.read_csv(student_data_path)

#Map the String values for each column needed
student_df['Parental_Involvement'] = student_df['Parental_Involvement'].map({'Low': 1, 'Medium': 2, 'High': 3})
student_df['Access_to_Resources'] = student_df['Access_to_Resources'].map({'Low': 1, 'Medium': 2, 'High': 3})
student_df['Motivation_Level'] = student_df['Motivation_Level'].map({'Low': 1, 'Medium': 2, 'High': 3})
```

```python
#Dataframe sorted by Exam sxocres over 90
A_grade_dataframe = student_df[(student_df["Exam_Score"] >= 90)]
A_grade_dataframe = A_grade_dataframe.sort_values(["Exam_Score"], ascending=False)
A_grade_dataframe.head(15)
```

```python
#Private and public dataframes
private_school_dataframe = student_df[(student_df["School_Type"] == "Private")]
public_school_dataframe = student_df[(student_df["School_Type"] == "Public")]
```

```python
#add exam score difference column
student_df['Score_Difference'] = student_df['Exam_Score'] - student_df['Previous_Scores']
grade_change_dataframe = student_df.sort_values(["Score_Difference"], ascending=False)
grade_change_dataframe.head(15)
```

```python
#dataframe sorted by Attendance
Sorted_dataframe = student_df.sort_values(["Attendance"], ascending=False)
Sorted_dataframe.head(15)
```
