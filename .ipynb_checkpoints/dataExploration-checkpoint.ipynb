```python
import pandas as pd
import numpy as np
```

```python
# Student data filepath
student_data_path = "data/StudentPerformanceFactors.csv"

# Student data dataframe created by reading the CSV with pandas
csv_df = pd.read_csv(student_data_path)

#prints number of students
print(f"This data set has {len(csv_df)} rows of student data")
csv_df.head()
```

```python
#Searches for any null values in data
csv_df.isnull().sum()
```

```python
#Drops null values can checks again
student_df = csv_df.dropna()
student_df.isnull().sum()
```

```python
#Removes the outlier score of 101
student_df = student_df[student_df["Exam_Score"] <= 100]
student_df.sort_values(["Exam_Score"], ascending=False)
```

```python
#prints number of students
print(f"This data set has {len(student_df)} rows of student data")
student_df.head()
```

```python
#add exam score difference column
student_df['Score_Difference'] = student_df['Exam_Score'] - student_df['Previous_Scores']
grade_change_dataframe = student_df.sort_values(["Score_Difference"], ascending=False)
grade_change_dataframe.head(15)
```

```python
#add exam score quartile placement column
#arrange the data in ascending order, then find the median of the lower half of the data
student_df['Quartile'] = pd.qcut(student_df['Exam_Score'], 4, labels=[1, 2, 3, 4])

q1 = student_df[student_df['Quartile'] == 1].sort_values(["Exam_Score"], ascending=False)
q3 = student_df[student_df['Quartile'] == 3].sort_values(["Exam_Score"], ascending=False)
iqr = q3['Exam_Score'] - q1['Exam_Score']

lower_half = q1.iloc[:len(q1) // 2]
median_lower_half = lower_half['Exam_Score'].median()
print("Median of the lower half for Q1:", median_lower_half)

lower_half = q3.iloc[:len(q3) // 2]
median_lower_half = lower_half['Exam_Score'].median()
print("Median of the lower half for Q3:", median_lower_half)

#Value Test: Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = student_df['Exam_Score'].quantile(0.25)
Q3 = student_df['Exam_Score'].quantile(0.75)

#Calculate the IQR
IQR = Q3 - Q1

#Display the results
print("Value Test")
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
```

```python
contingency_table = pd.crosstab(student_df['Parental_Involvement'], student_df['Motivation_Level'])
# Define the custom order for rows and columns
row_order = ['Low', 'Medium', 'High']
column_order = ['Low', 'Medium', 'High']
# Sort the contingency table by the custom order for rows and columns
contingency_table = contingency_table.reindex(index=row_order, fill_value=0)
contingency_table = contingency_table.reindex(columns=column_order, fill_value=0)
# Print the sorted contingency table
print(contingency_table)
```

```python
#output csv of the finsihed df for data analysis notebook
student_df.to_csv("data/studentPerformanceFactors_Cleaned.csv", index = False)
```
